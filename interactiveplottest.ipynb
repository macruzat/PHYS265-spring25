```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

L, n = 2, 400
x = np.linspace(-L,L,n)
y = x.copy()
X, Y = np.meshgrid(x,y)
Z = np.exp(-(X**2+Y**2))

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_wireframe(X,Y,Z,rstride=40,cstride=40)
```
